```python
import pandas as pd

econ = pd.read_csv("../data/raw/socio_economic_economic.csv")
demo = pd.read_csv("../data/raw/socio_economic_demographic.csv")
law  = pd.read_csv("../data/raw/socio_economic_law_order.csv")
crime = pd.read_csv("../data/raw/crime_data.csv")

print("Economic Columns:", econ.columns.tolist())
print("Demographic Columns:", demo.columns.tolist())
print("Law Columns:", law.columns.tolist())
print("Crime Columns:", crime.columns.tolist())

```

```python
print(econ.duplicated(subset=['State','Year']).sum())
print(demo.duplicated(subset=['State','Year']).sum())
print(law.duplicated(subset=['State','Year']).sum())
```

```python
socio = econ.merge(demo, on=["State", "Year"], how="inner")
socio = socio.merge(law, on=["State", "Year"], how="inner")
final = crime.merge(socio, on=["State", "Year"], how="inner")

final.to_csv("../data/cleaned/final_merged_dataset.csv", index=False)
print("Final dataset ready and saved")
```
